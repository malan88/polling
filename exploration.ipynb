```python
import pandas as pd

df = pd.read_csv('polls.csv')
df.head()
```

```python
national_polls = df[df['state'].isnull()]
national_pollsters = national_polls['pollster'].unique()
```

```python
national_polls.columns
```

```python
candidates = national_polls.loc[:,'candidate_name'].unique()[0:2]
candidates
```

```python
national_polls[national_polls['candidate_name'].isin(candidates)].shape
```

```python
national_polls.shape
```

```python
national_polls = national_polls[national_polls['candidate_name'].isin(candidates)]
```

```python
national_polls[national_polls['pollster'] == national_pollsters[0]]
```

```python
date_cols = ['start_date', 'end_date', 'election_date','created_at',]
for col in date_cols:
    national_polls[col] = pd.to_datetime(df[col])
```

```python
national_polls[col]
```

```python
#national_polls['diff'] = national_polls.groupby('question_id')['pct'].apply(lambda x: x.shift(-1) - x)
national_polls['diff'] = national_polls.groupby('question_id')['pct'].diff(-1)
```

```python
npolls = national_polls
npolls[(npolls['pollster'] == national_pollsters[0]) & npolls['diff'].notnull()]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='end_date', y='diff', data=npolls[(npolls['pollster'] == national_pollsters[0]) & npolls['diff'].notnull()]).set_title('Ipsos')
```

```python
for i, pollster in enumerate(national_pollsters):
    if npolls[(npolls['pollster'] == pollster) & npolls['diff'].notnull()].shape[0] > 1:
        plt.figure(i)
        plot = sns.lineplot(x='end_date', y='diff', data=npolls[(npolls['pollster'] == pollster) & npolls['diff'].notnull()]).set_title(pollster)
        fig = plot.get_figure()
        fig.savefig('national/' + f'{pollster}.png'.replace('/', ''))
```
